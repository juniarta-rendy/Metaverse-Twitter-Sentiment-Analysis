```python
#swear lexicon dictionary
file = open('../../Krispi/Lexicon/Lexicon Dictionary/swearwords/swear-words.txt','r')
content2 = file.read()
swear_words = content.split("\n")
```

```python
weight_swear = [-5 for i in range(len(swear_words))]
```

```python
swear_lexicon = pd.DataFrame(list(zip(swear_words,weight_swear)),columns =['word','weight'])
```

```python
full_lexicon = lexicon.append(swear_lexicon, ignore_index = True)
```

```python
#Tokenize Text
def number_of_words(x):
    words = word_tokenize(x['word'])
    number = len(words)
    return number
```

```python
full_lexicon['number_of_words'] = full_lexicon.apply(lambda x:number_of_words(x),axis=1)
```

```python
#create Stemmer
factory = StemmerFactory()
Stemmer = factory.create_stemmer()
```

```python
#import stopwords dictionary
my_file = open('../../Krispi/Lexicon/Lexicon Dictionary/stopwords/combined_stop_words.txt','r')
content = my_file.read()
stop_words = content.split('\n')
```

```python
lexicon_word = modified_lexicon['word'].tolist()
```

```python
overlap=[]
for word in stop_words:
    if word in lexicon_word:
        overlap.append(word)
    else:
        kata_dasar = Stemmer.stem(word)
        if kata_dasar in lexicon_word:
            overlap.append(word)
```
